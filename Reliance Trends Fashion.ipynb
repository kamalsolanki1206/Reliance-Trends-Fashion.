```python
data.dropna(inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="Brand",data=data)
plt.title("CountPlot as Brand")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
data["Brand"]=lb.fit_transform(data["Brand"])
```

```python
data.head()
```

```python
data.info()
```

```python
data["Description"]=lb.fit_transform(data["Description"])
data["Image_URL"]=lb.fit_transform(data["Image_URL"])
data["Discount_Price (in Rs.)"]=lb.fit_transform(data["Discount_Price (in Rs.)"])
data["Original_Price (in Rs.)"]=lb.fit_transform(data["Original_Price (in Rs.)"])
data["Product_URL"]=lb.fit_transform(data["Product_URL"])
data["Category_by_gender"]=lb.fit_transform(data["Category_by_gender"])
data["Category"]=lb.fit_transform(data["Category"])
```

```python
data.head()
```

```python
sns.histplot(x="Brand",data=data)
plt.title("CountPlot as Brand")
plt.show()
```

```python
sns.histplot(x="Description",data=data)
plt.title("CountPlot as Description")
plt.show()
```

```python
sns.histplot(x="Discount_Price (in Rs.)",data=data)
plt.title("CountPlot as Discount_Price (in Rs.)")
plt.show()
```

```python
sns.histplot(x="Original_Price (in Rs.)",data=data)
plt.title("CountPlot as Original_Price (in Rs.)")
plt.show()
```

```python
sns.countplot(x="Category_by_gender",data=data)
plt.title("CountPlot as Category_by_gender")
plt.show()
```

```python
sns.countplot(x="Category",data=data)
plt.title("CountPlot as Category")
plt.show()
```
